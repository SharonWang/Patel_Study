```python
#%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
import scanpy as sc
import pandas as pd
from os import listdir
from os.path import isfile, join
import re
import anndata
import seaborn as sns

plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

sc.settings.verbosity = 3  # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()

from matplotlib.colors import LinearSegmentedColormap
cmap = LinearSegmentedColormap.from_list(name='gene_cmap', colors=['lightgrey', 'thistle', 'red', 'darkred']) 

sc.settings.set_figure_params(dpi=80, color_map='viridis', vector_friendly=False,  dpi_save=300)
```

```python
adata = sc.read('./write/CD34_raw_afterQC_all_lognorm_hvg_processed_harmony_filtered.h5ad')
```

```python
p53GL = np.genfromtxt('/home/xw251/rds/rds-bg200-hphi-gottgens/users/xw251/Lab_Members/Niki/Patel_SMQ2/Analysis_17Feb2021/p53GL.txt', dtype='str')
print(len(p53GL))
p53GL = [x.upper() for x in p53GL]
p53GL = np.intersect1d(p53GL, adata.raw.var_names)
```

```python
p53GL
```

```python
adata_norm = anndata.AnnData(X=adata.raw.X, var=adata.raw.var, obs=adata.obs, obsm=adata.obsm)
```

```python
sc.pp.scale(adata_norm)
```

```python
sc.tl.score_genes(adata_norm, p53GL, score_name = 'p53S', use_raw=False)
```

```python
adata.obs['p53S'] = adata_norm.obs['p53S']
```

```python
# FigureS5E
from statannot import add_stat_annotation
import seaborn as sns
cat = adata.obs['CT_new'].cat.categories
fig, ax = plt.subplots(3,4, figsize=(12,8), squeeze=False, sharex=True)
ax = ax.flatten()
for i in range(len(cat)):
    print(cat[i])
    idx = adata.obs['CT_new'] == cat[i]
    df = adata[idx,:].obs[['p53S', 'Condition']]
    print(df['Condition'].cat.categories)

    sns.violinplot(x='Condition', y='p53S', scale='width', data=df,ax=ax[i])
    sns.stripplot(x="Condition", y="p53S", data=df,ax=ax[i], jitter=True, zorder=1, color=".3")
    ax[i].grid(False)
    ax[i].set_title(cat[i])
    ax[i].set_xlabel('')
    ax[i].set_ylabel('')
    ax[i].axhline(0, linestyle='--', color='grey')
    ax[i].set_xticklabels(labels=df['Condition'].cat.categories,rotation=90)
    if len(df['Condition'].cat.categories) > 1:
        add_stat_annotation(ax[i], data=df, x='Condition', y='p53S',
                            box_pairs=[('Healthy', 'Patient')],comparisons_correction=None,
                            test='Mann-Whitney', text_format='simple', loc='inside', verbose=2)

# fig.text(0.5, -0.01, 'Age', ha='center')
fig.text(-0.01, 0.5, 'p53 score', ha='center', rotation='vertical')

# for i in [15]:
#     fig.delaxes(ax.flatten()[i])
    
plt.tight_layout(pad=0.5)
```
