```r
library("TCGAbiolinks")
library(dplyr)
```

```r
GDCprojects = getGDCprojects()
```

```r
######### TCGA LAML
query_TCGA = GDCquery(
  project = "TCGA-LAML",
  data.category = "Transcriptome Profiling", # parameter enforced by GDCquery
  experimental.strategy = "RNA-Seq",
  workflow.type = "STAR - Counts",
  data.type = "Gene Expression Quantification") # 151

GDCdownload(query = query_TCGA)
tcga_data = GDCprepare(query_TCGA)
saveRDS(tcga_data, file="TCGA_LAML.rds")
```

```r
library(SummarizedExperiment)
```

```r
saveRDS(assay(tcga_data), file="TCGA_LAML_rawcnt.rds")
saveRDS(colData(tcga_data), file="TCGA_LAML_coldata.rds")
saveRDS(rowData(tcga_data), file="TCGA_LAML_rowdata.rds")
```

```r
cnt = readRDS("TCGA_LAML_rawcnt.rds")
```

```r
library(edgeR)
```

```r
y = DGEList(counts = cnt)
y = calcNormFactors(y)
logcpm = cpm(y, log=TRUE)
saveRDS(logcpm, file='TCGA_LAML_logcpm.rds')
```

```r
geneInfo = readRDS("TCGA_LAML_rowdata.rds")
logcpm = readRDS("TCGA_LAML_logcpm.rds")
rownames(logcpm) = geneInfo$gene_name
```

```r
p53GL = read.table("/home/xw251/rds/rds-bg200-hphi-gottgens/users/xw251/Lab_Members/Niki/Patel_SMQ2/Analysis_17Feb2021/p53GL.txt")[,1]
p53Gexp = colMeans(logcpm[rownames(logcpm) %in% toupper(p53GL), ])
```

```r
meta = readRDS("TCGA_LAML_coldata.rds")
meta$os<-ifelse(meta$vital_status=='Alive',meta$days_to_last_follow_up,meta$days_to_death)
meta[meta$vital_status=='Dead',]$vital_status <- 2
meta[meta$vital_status=='Alive',]$vital_status <- 1
meta$vital_status <- as.numeric(meta$vital_status)
meta$p53Gexp = p53Gexp
```

```r
print(dim(meta))
```

```r
# meta = meta[!is.na(meta$os),]
# meta = meta[!is.na(meta$vital_status),]
# meta = meta[meta$os <= 2000,]
```

```r
print(dim(meta))
```

```r
cutoffs = quantile(meta$p53Gexp, c(0.25, 0.75))

meta$p53_class = rep("p53_mid", nrow(meta))
meta$p53_class[meta$p53Gexp < cutoffs[1]] = "p53_low"
meta$p53_class[meta$p53Gexp > cutoffs[2]] = "p53_high"
print(table(meta$p53_class))
```

```r
library(survival)
library(survminer)
```

```r
fit = survfit(Surv(os, vital_status)~p53_class, data=meta)
```

```r
# FigureS6B - all patients
ggsurvplot(fit, pval=TRUE, palette = c('red', 'green', 'blue'),
           legend.title="p53genes", risk.table=TRUE,
           tables.height=0.2, tables.theme=theme_cleantable(),
           xlab='Time(days)')
```

```r
# FigureS6E - all patients
# Prior treatment removed
meta_np = meta[meta$prior_treatment == "No",]
print(dim(meta_np))

cutoffs = quantile(meta_np$p53Gexp, c(0.25, 0.75))

meta_np$p53_class[meta_np$p53Gexp <= cutoffs[2]] = "p53_low"
meta_np$p53_class[meta_np$p53Gexp > cutoffs[2]] = "p53_high"
print(table(meta_np$p53_class))

fit = survfit(Surv(os, vital_status)~p53_class, data=meta_np)
ggsurvplot(fit, pval=TRUE, palette = c('red', 'green'),
           legend.title="p53genes", risk.table=TRUE,
           tables.height=0.2, tables.theme=theme_cleantable(),
           xlab='Time(days)')
```

```r
# FigureS6E - all patients
# Prior treated
meta_p = meta[meta$prior_treatment == "Yes",]
print(dim(meta_p))

cutoffs = quantile(meta_p$p53Gexp, c(0.25, 0.75))

meta_p$p53_class[meta_p$p53Gexp <= cutoffs[2]] = "p53_low"
meta_p$p53_class[meta_p$p53Gexp > cutoffs[2]] = "p53_high"
print(table(meta_p$p53_class))

fit = survfit(Surv(os, vital_status)~p53_class, data=meta_p)
ggsurvplot(fit, pval=TRUE, palette = c('red', 'green'),
           legend.title="p53genes", risk.table=TRUE,
           tables.height=0.2, tables.theme=theme_cleantable(),
           xlab='Time(days)')
```

```r
# FigureS6E - all patients
# Age > 60
meta_old = meta[meta$age_at_index > 60,]
print(dim(meta_old))

cutoffs = quantile(meta_old$p53Gexp, c(0.25, 0.75))

meta_old$p53_class[meta_old$p53Gexp <= cutoffs[2]] = "p53_low"
meta_old$p53_class[meta_old$p53Gexp > cutoffs[2]] = "p53_high"
print(table(meta_old$p53_class))

fit = survfit(Surv(os, vital_status)~p53_class, data=meta_old)
ggsurvplot(fit, pval=TRUE, palette = c('red', 'green'),
           legend.title="p53genes", risk.table=TRUE,
           tables.height=0.2, tables.theme=theme_cleantable(),
           xlab='Time(days)')
```

```r
# FigureS6E - all patients
# Age < 60
meta_young = meta[meta$age_at_index < 60,]
print(dim(meta_young))

cutoffs = quantile(meta_young$p53Gexp, c(0.25, 0.75))

meta_young$p53_class[meta_young$p53Gexp <= cutoffs[2]] = "p53_low"
meta_young$p53_class[meta_young$p53Gexp > cutoffs[2]] = "p53_high"
print(table(meta_young$p53_class))

fit = survfit(Surv(os, vital_status)~p53_class, data=meta_young)
ggsurvplot(fit, pval=TRUE, palette = c('red', 'green'),
           legend.title="p53genes", risk.table=TRUE,
           tables.height=0.2, tables.theme=theme_cleantable(),
           xlab='Time(days)')
```

```r
library(ggpubr)
```

```r
# FigureS6C - all patients
ggscatter(as.data.frame(meta[,c("age_at_index", "p53Gexp")]), x = "age_at_index", y = "p53Gexp",
          size =2, # Points color and size
          add = "reg.line",  # Add regression line
          add.params = list(color = "grey", fill = "orange"), # Customize regression line
          conf.int = TRUE, # Add confidence interval
          cor.coef = TRUE, # Add correlation coefficient. see ?stat_cor
          cor.coeff.args = list(method = "pearson"),
          xlab = "Age", ylab="p53 exp")
```

```r
df = meta[,c("p53Gexp", "age_at_index")]
df$Age = ifelse(df$age_at_index>60, ">60", "<60")
```

```r
# FigureS6C
ggviolin(as.data.frame(df), x="Age", y="p53Gexp", add = c("jitter", "boxplot"), fill="Age", add.params = list(fill = "white")) +
    stat_compare_means(comparisons=list(c("<60", ">60")), method="t.test")
```

```r
meta$Aldh2exp = logcpm[grep("ALDH2", rownames(logcpm)),]
```

```r
df = meta[,c("Aldh2exp", "p53_class")]
df = meta[meta$p53_class != "p53_mid",]
```
